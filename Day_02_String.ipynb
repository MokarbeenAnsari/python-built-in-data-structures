#### **Python String**

**Strings** in python are surrounded by either **single** quotation marks, or **double** quotation marks.

`'Hello Python'` is the same as `"Hello Python"`.

````python
# Printing the string
print("Hello Python")
print('Hello Python')
````

#### **Assign String to Variable**

We can store the string in a varibale using assignment operator.

````python
# Assignment of string
a = "Hello Python"
b = 'Hello Python'

print(a)
print(b)
````

#### **Multi-line String**

We can assign multi-line string to variable using **triple** single/double quotation marks.

````python
a = """This is line1.
This is line2.
This is line3.
This is line4.
This is line5."""

b = '''This is line1.
This is line2.
This is line3.
This is line4.
This is line5.'''

print(a)
print(b)
````

#### **Special Note**

During **string** creation always use **same** quotation mark for opening and closing.

````python
a = "Mokarbeen Ansari'
print(a)
````

#### **Strings are Arrays**

Like many other popular programming languages, strings in Python are arrays of bytes representing **unicode** characters.

However, Python does not have a **character data type**, a single character is simply a **string with a length of 1**.

**Square brackets** can be used to access elements of the string.

````python
a = "Hello World!"

# Indexing from beginning start from 0
print(a[0])
print(a[1])
print("---")
# indexing from end start from -1
print(a[-1])
print(a[-2])
````

#### **Looping Through a String**

Since strings are arrays, we can loop through the characters in a string, with a for loop.

````python
# Example
a = "Hello World!"
for x in a:
  print(x)
````

#### **String Length**

To get the length of a string, use the len() function.

````python
# Example:
a = 'Mokarbeen Ansari'
len(a)
````

#### **Check in String**

To check if a certain phrase or character is present in a string, we can use the keyword **`in`**.

````python
# Example
text = "This is Mokarbeen, these days I am learning AI."
print("Mokarbeen" in text)
print("Mokarbeen Ansari" in text)
````

````python
# Example
text = "This is Mokarbeen, these days I am learning AI."
name = "Mokarbeen"

if name in text:
  print("Learning Artificial intelligence!")
````

````python
# Example
text = "This is Mokarbeen, these days I am learning AI."
name = "Mokarbeen Ansari"

if name not in text:
  print("Learning other than Artificial intelligence!")
````

#### **String Index Slicing**

You can return a range of characters by using the slice syntax.

Specify the start index and the end index, separated by a colon, to return a part of the string.

```python
String[start : stop : step]
```



````python
# Slicing from beginning
a = 'Mokarbeen Ansari'
print(a)
print(a[0:])
print(a[0:5])
print(a[:5])
print(a[3:7])
````

````python
# Slicing from end
a = 'Mokarbeen Ansari'
print(a[-1:])
print(a[-6:-1])
print(a[-1:-5:-1])
print(a[-1::-1])       # String reverse
````

#### **String Methods**

- **`capitalize()`** - 	Converts the first character to upper case

````python
name = "mokarbeen"
print(name.capitalize())
````

- **`upper()`**	  - Converts a string into upper case 
- **`isupper()`** - Returns True if all characters in the string are upper case

````python
name = "Mokarbeen Ansari"
print(name.isupper())

print(name.upper())
print(name.upper().isupper())
````

- **`lower()`**	- Converts a string into lower case
- **`islower()`**	- Returns True if all characters in the string are lower case

````python
name = 'Mokarbeen Ansari'
print(name.islower())
print(name.lower())
print(name.lower().islower())
````

- **`rstrip()`**	- Returns a right trim version of the string
- **`lstrip()`**	- Returns a left trim version of the string
- **`strip()`**	  - Returns a trimmed version of the string
- **`isspace()`**	- Returns True if all characters in the string are whitespaces

````python
name = " "

print(name.isspace())

name = "   Mokarbeen Ansari  "
print(name)
print(name.lstrip())
print(name.rstrip())
print(name.strip())
````

- **`replace()`**	- Returns a string where a specified value is replaced with a specified value

````python
a = "Hello World!"
print(a.replace("World", "Mokarbeen" ))
````

- **`split()`**	- Splits the string at the specified separator, and returns a list

````python
name = "Mokarbeen, Reshma, Afiya, Ayan"
print(name.split(','))
````

- **`title()`** - Converts the first character of each word to upper case
- **`istitle()`** - Returns True if the string follows the rules of a title

````python
name = 'md mokarbeen ansari'

print(name.istitle())

print(name.title())
print(name.title().istitle())
````

- **`count()`**	- Returns the number of times a specified value occurs in a string

````python
name = "Md Mokarbeen Ansari"
print(name.count('M'))
print(name.count('a'))
print(name.count('c'))
````

- **`find()`**-	Searches the string for a specified value and returns the position of where it was found
- **`rfind()`** -	Searches the string for a specified value and returns the last position of where it was found
- **`index()`**	-Searches the string for a specified value and returns the position of where it was found
- **`rindex()`** -	Searches the string for a specified value and returns the last position of where it was found

````python
name = 'Md Mokarbeen Ansari'
print(name.find('r'))
print(name.index('r'))
````

- **`isalnum()`**-	Returns True if all characters in the string are alphanumeric
- **`isalpha()`**-	Returns True if all characters in the string are in the alphabet
- **`isdecimal()`**-	Returns True if all characters in the string are decimals
- **`isdigit()`**- Returns True if all characters in the string are digits
- **`isidentifier()`**-	Returns True if the string is an identifier
- **`isnumeric()`**-	Returns True if all characters in the string are numeric

- **`join()`**-	Converts the elements of an iterable into a string

````python
my_tuple = ('Md', 'Mokarbeen', 'Ansari')
my_list = ['Md', 'Mokarbeen', 'Ansari']
my_dict = {'0': 'Md', '1':'Mokarbeen', '2':'Ansari'}
print(' '.join(my_dict.values()))
print('#'.join(my_tuple))
````

#### **String Concatenation**
To concatenate, or combine, two or more strings we can use the + operator.

````python
a = 'Mokarbeen'
b = 'Ansari'
name = a + ' ' + b
print(name)
````

#### **Format String**

````python
name = 'Mokarbeen'
age = 37

print("My name is {0} and I am {1} years old.".format(name, age))
print(f"My name is {name} and I am {age} years old.")
````

#### **Escape Character**

To insert characters that are illegal or change the original meaning in a string, use an escape character.

An escape character is a backslash \ followed by the character you want to insert.

An example of an illegal character is a double quote inside a string that is surrounded by double quotes:

````python
name = "Mokarbeen Ansari"
print(f"My name is \"{name}\".")
````
